Data Visualization using Matplotlib and Seaborn


```python
import pandas as pd

data = pd.read_csv('/content/computer_sales.csv')
```

```python
data.sample()
```

```python
#Numeric and non-numeric data
#Numeric is either continuous(90,98,70,89) or discrete data (semester 1,2,3,4,5,6)
#Non-numeric data is called categoric data
#
```

```python
import matplotlib.pyplot as plt
import matplotlib_inline
```

```python
plt.plot(data['Contact Age'])
```

```python
plt.plot(data['Sale Price'])
```

```python
plt.plot(data['Profit'])
```

```python
plt.plot(data['Contact Age'])
plt.plot(data['Profit'])
plt.plot(data['Our Cost'])
plt.plot(data['Sale Price'])
plt.show()
```

```python
plt.hist(data['Profit'])
```

```python
plt.pie(data['Profit'])
```

```python
data['Profit'].plot(kind='bar',color='green')
```

```python
data['Sale ID'].plot(kind='bar',color='green')
```

```python
#Bivariate data analysis
plt.scatter(x='Sale Price',
            y='Profit',
            data=data)
```

```python
data.corr()
```

```python
correlation = data.corr()
```

```python
plt.imshow(correlation)
```

```python
import seaborn as sns
```

```python
sns.lineplot(x='Sale ID',
             y='Profit',
             data=data)
```

```python
sns.barplot(x='Sale ID',
             y='Profit',
             data=data)
```

```python
sns.heatmap(correlation)
```

```python
sns.heatmap(correlation,annot=True)
```

```python
sns.countplot(x='Contact City', data=data)
```

```python
sns.pairplot(hue=)
```

```python
sns.regplot(x='Sale Price',
            y='Profit',
            data=data)
```

```python
sns.regplot(x='Our Cost',
            y='Profit',
            data=data)
```

```python
sns.lmplot(x='Sale Price',
           y='Profit',
           data=data,
           col='Contact City',
           hue='Contact City')
```

```python
sns.boxplot(x='Contact City',
            y='Profit',
            data=data)
```

```python
sns.violinplot(x='Contact City',
               y='Profit',
               data=data)
```

```python
sns.stripplot(x='Contact City',
              y='Profit',
              data=data)
```
