```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
data = sns.load_dataset('car_crashes') #Importing anagrams dataset
```

```python
data.info() #checking out information in data
```

```python
data.head()
```

```python
print("Minimum of car crashes data", data.min()),'\n','\n'
print("Maximum of car crashes data", data.max()),'\n','\n'
print("Mean of car crashes data", data.mean()),'\n','\n'
print("Median of car crashes data", data.median()),'\n','\n'
print("Variance of car crashes data", data.var()),'\n','\n'
print("Standard deviation of car crashes data", data.std()),'\n','\n'
```
